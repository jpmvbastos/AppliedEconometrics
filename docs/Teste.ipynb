```python
# Step 2: Implement Data Filtering
def filter_data(uf, candidate, cargo):
    filtered_df = df[(df['SG_UF'] == uf) & (df['NM_CANDIDATO'] == candidate) & (df['DS_CARGO'] == cargo)]
    return filtered_df

# Step 3: Create an Interactive Interface with Dynamic Options
uf_options = sorted(df['SG_UF'].unique().tolist())

uf_dropdown = Dropdown(options=uf_options)
candidate_dropdown = Dropdown()
cargo_dropdown = Dropdown(options=sorted(df['DS_CARGO'].unique().tolist()))

def update_candidates(*args):
    selected_uf = uf_dropdown.value
    candidate_options = sorted(df[df['SG_UF'] == selected_uf]['NM_CANDIDATO'].unique().tolist())
    candidate_dropdown.options = candidate_options

uf_dropdown.observe(update_candidates, 'value')

@interact
def interactive_ide(uf=uf_dropdown,
                    candidate=candidate_dropdown,
                    cargo=cargo_dropdown):
    filtered_data = filter_data(uf, candidate, cargo)
    display(filtered_data)
```
